```python
import pandas as pd
import matplotlib.pyplot as plt

# Data extracted from the table
data = {
    'Age Group': ['20s', '30s', '40s', '50s', '60s'],
    'TV': [4.00, 4.09, 4.16, 4.24, 4.42],
    'Internet Portal': [4.04, 4.14, 4.11, 4.03, 4.02],
    'Social Media': [4.16, 4.08, 3.98, 3.86, 3.84],
    'Internet Newspaper': [3.72, 3.82, 3.80, 3.75, 3.72],
    'Radio': [3.03, 3.38, 3.54, 3.62, 3.67],
    'Print Newspaper': [2.99, 3.23, 3.26, 3.38, 3.64]
}

df = pd.DataFrame(data)

# Calculating the average ratings for age groups 30s and above
df_older = df[df['Age Group'] != '20s']
average_ratings_older = df_older.mean()

# Extracting ratings for 20s
ratings_20s = df[df['Age Group'] == '20s'].iloc[0].drop('Age Group')

# Calculating differences
differences = ratings_20s - average_ratings_older

# Color mapping: Positive differences in blue, negative in red
colors = ['blue' if x > 0 else 'red' for x in differences]

# Plotting the differences
plt.figure(figsize=(10, 6))
differences.plot(kind='bar', color=colors)
plt.title('Difference in Media Ratings between 20s and Older Age Groups')
plt.xlabel('Media Type')
plt.ylabel('Rating Difference')
plt.axhline(0, color='gray', linewidth=0.8)  # Adds a line at zero difference for reference
plt.xticks(rotation=45)
plt.show()
```
